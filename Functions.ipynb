```python
states = ["Alabama", "Georgia!", "Gerorgia", "georgia", "FlOrIda", "South carolina##", "West virginia?"]
```

```python
import re
```

```python
def clean_strings(strings):
  result = []
  for value in strings:
    value = value.strip()
    value = re.sub("[!#?]", "", value)
    value = value.title()
    result.append(value)
  return result
```

```python
clean_strings(states)
```

```python
def squares(n=10):
  print(f"Generating squares from 1 to {n ** 2}")
  for i in range(1, n + 1):
    yield i ** 2
```

```python
gen = squares()
gen
```

```python
for x in gen:
  print(x, end=" ")
```

```python
import itertools
```

```python
def first_letter(x):
  return x[0]
```

```python
names = ["Alan", "Adam", "Wes", "Will", "Albert", "Steven"]
for letter, names in itertools.groupby(names, first_letter):
  print(letter, list(names))
```
