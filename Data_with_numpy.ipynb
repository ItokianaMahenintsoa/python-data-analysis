```python
names = np.array(["Bob", "Joe", "Will", "Bob","Will", "Joe", "Joe"])
data = np.array([[4, 7], [0, 2], [-5, 6], [0, 0], [1,2], [-12, -4], [3, 4]])
```

```python
names
```

```python
data
```

```python
names == "Bob"
```

```python
data[names == "Bob"]
```

```python
data[names == "Bob", 1:]
```

```python
data[names == "Bob", 1]
```

```python
names != "Bob"
```

```python
~(names == "Bob")
```

```python
data[~(names == "Bob")]
```

```python
cond = names == "Bob"
data[~cond]
```

```python
mask = (names == "Bob") | (names == "Will")
mask
```

```python
data[mask]
```

```python
data[data < 0] = 0
data
```

```python
data[names != "Joe"] = 7
data
```
