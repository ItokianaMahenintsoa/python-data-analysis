```python
sink_path = "sink.txt"
```

```python
with open(path) as source:
  with open(sink_path, "x",  encoding="iso-8859-1") as sink:
    sink.write(source.read())
```

```python
with open(sink_path, encoding="iso-8859-1") as f:
  print(f.read(10))
```
